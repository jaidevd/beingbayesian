```python
import pandas as pd
```

```python
edu = pd.read_parquet('data/edu-indicators.parquet')
edu.head()
```

```python
edu.columns
```

```python
edu['is_hhmem_edu'].value_counts(normalize=True)
```

```python
edu['received_fee_waiver'].unique()
```

```python
has_edu = edu['is_hhmem_edu'] == 1
edu[~has_edu].drop(['is_hhmem_edu'], axis=1).isnull().sum(axis=1).unique()
```

```python
for col in edu.drop(['is_hhmem_edu'], axis=1):
    print(col, edu[col].unique())
```

```python
def edu_score(row):
    if row['is_hhmem_edu'] == 2:
        return 0
    score = row[['received_free_textbook', 'received_free_stationary', 'received_free_schoolbag', 'received_free_oth_school']]
    score = score.sum() / 5
    return score + (row['received_fee_waiver'] == 1).astype(int) / 5

edu['edu_score'] = edu.apply(edu_score, axis=1)
```

```python
edu.to_parquet('data/edu-indicators.parquet')
```
