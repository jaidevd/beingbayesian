```python
from glob import glob
import pandas as pd
```

```python
filepatterns = [f"../../hces-2023-24/data/sec-{i}.*.parquet" for i in range(5, 15)]
cons_files = []
for pat in filepatterns:
    cons_files.extend(glob(pat))
```

```python
consumption, sources = [], []

for file in cons_files:
    df = pd.read_parquet(file)
    colname = 'cons_total_value' if 'cons_total_value' in df else 'tot_expenditure'
    consumption.append(df[colname].unstack())
    if 'source' in df:
        sources.append(df['source'].unstack())
```

```python
cdf = pd.concat(consumption, axis=1, verify_integrity=True)
cdf.head()
```

```python
sdf = pd.concat(sources, axis=1, verify_integrity=True)
sdf.head()
```

```python
cdf.to_parquet('data/baskets-consumption-value.parquet')
sdf.to_parquet('data/baskets-sources.parquet')
```
