```python
cereals = pd.read_parquet('../../hces-2023-24/data/sec-5.1-FDQ.parquet')
cereals = cereals['cons_total_value'].unstack()
cereals = cereals.drop([129], axis=1).sum(axis=1)
```

```python
pulses = pd.read_parquet('../../hces-2023-24/data/sec-5.2-FDQ.parquet')
pulses = pulses['cons_total_value'].unstack()
pulses = pulses.drop([159], axis=1).sum(axis=1)
```

```python
ss = pd.read_parquet('../../hces-2023-24/data/sec-5.3-FDQ.parquet')
ss = ss['cons_total_value'].unstack()
ss = ss.drop([179], axis=1).sum(axis=1)
```

```python
milk = pd.read_parquet('../../hces-2023-24/data/sec-6.1-FDQ.parquet')
milk = milk['cons_total_value'].unstack()
milk = milk.drop([169], axis=1).sum(axis=1)
```

```python
veg = pd.read_parquet('../../hces-2023-24/data/sec-6.2-FDQ.parquet')
veg = veg['cons_total_value'].unstack()
veg = veg.drop([219], axis=1).sum(axis=1)
```

```python
fruits = pd.read_parquet('../../hces-2023-24/data/sec-6.3-FDQ.parquet')
fruits = fruits['cons_total_value'].unstack()
fruits = fruits.drop([239], axis=1).sum(axis=1)
```

```python
meat = pd.read_parquet('../../hces-2023-24/data/sec-6.5-FDQ.parquet')
meat = meat['cons_total_value'].unstack()
meat = meat.drop([199], axis=1).sum(axis=1)
```

```python
oil = pd.read_parquet('../../hces-2023-24/data/sec-6.6-FDQ.parquet')
oil = oil['cons_total_value'].unstack()
oil = oil.drop([189], axis=1).sum(axis=1)
```

```python
spice = pd.read_parquet('../../hces-2023-24/data/sec-6.7-FDQ.parquet')
spice = spice['cons_total_value'].unstack()
spice = spice.drop([269], axis=1).sum(axis=1)
```

```python
bev = pd.read_parquet('../../hces-2023-24/data/sec-6.8-FDQ.parquet')
bev = bev['cons_total_value'].unstack()
bev = bev.drop([279], axis=1).sum(axis=1)
```

```python
cooked = pd.read_parquet('../../hces-2023-24/data/sec-7.1-FDQ.parquet')
cooked = cooked['cons_total_value'].unstack()
cooked = cooked.drop([289], axis=1).sum(axis=1)
```

```python
proc = pd.read_parquet('../../hces-2023-24/data/sec-7.2-FDQ.parquet')
proc = proc['cons_total_value'].unstack()
proc = proc.drop([299], axis=1).sum(axis=1)
```

```python
food = pd.concat([
    cereals,
    pulses,
    ss,
    milk * 52 / 12,
    veg * 52 / 12,
    fruits * 52 / 12,
    meat * 52 / 12,
    oil * 52 / 12,
    spice * 52 / 12,
    bev * 52 / 12,
    cooked * 52 / 12,
    proc * 52 / 12
], axis=1, verify_integrity=True)
food.columns = 'cereals pulses sugar_salt dairy veg fruits meat oil spice bev cooked processed'.split()
```

```python
food.to_parquet('data/food.parquet')
```
